### 2. Projeção do discriminante linear de Fisher (LDA)

LDA é uma técnica de redução de dimensionalidade que tem como propósito a maximização da separação entre as médias das classes projetadas, reduzindo a variância inter-classe de cada classe, de modo a **minimizar** a sobreposição entre as classes. Consiste em projetar os dados sobre a direção **w** que maximiza o funcional $J(w)$ = $\dfrac{(\mu_1 - \mu_2)^2}{s^2_1 + s^2_2}$, onde $s^2_1 + s^2_2$ é a variância total intra-classe.

```python
X_binary_class = np.asarray(data.drop('y', axis=1))
y_binary_class = np.asarray(data['y'])
print(X_binary_class.shape)
print(y_binary_class.shape)
```

```python
C0 = []
C0_labels = []

C1 = []
C1_labels = []

for i in range(len(y_binary_class)):
    y = y_binary_class[i]
    x = X_binary_class[i]
    if y == 0:
        C0.append(x)
        C0_labels.append(y)
    elif y == 1:
        C1.append(x)
        C1_labels.append(y)
    
C0 = np.asarray(C0)
C0_labels = np.asarray(C0_labels)
C1 = np.asarray(C1)
C1_labels = np.asarray(C1_labels)

print(C1.shape, C1_labels.shape)
```

* Cálculo de $\mu_1$ e $\mu_2$:

```python
# Computa o vetor média para cada uma das 2 classes
m0 = np.mean(C0,axis=0)
m1 = np.mean(C1,axis=0)
print(m0, m1)
```

- Cálculo de $S^{-1}_w$

```python
# Cálculo de S_w 
tmp = np.subtract(C0, m0)
a = np.dot(tmp.T, tmp)

tmp = np.subtract(C1, m1)
b = np.dot(tmp.T, tmp)

SW = np.add(a,b)

inv_SW = np.linalg.inv(SW)
s = m1 - m0
W = np.dot(inv_SW, s) # 
print(W)
```

```python
# Vetor de predição
y_hat = np.dot(X_binary_class,W)
print(y_hat.shape)
#y_hat
```

- Projeção do discriminante: $w^* \propto S^{-1}_w (\mu_1 - \mu_2)$ 

```python
fig, ax = plt.subplots(figsize=(10,6), dpi=80, facecolor='w', edgecolor='k')
colors=['blue','red']
plt.plot(data_class_0['x_1'], data_class_0['x_2'], 'o', alpha=.6)
plt.plot(data_class_1['x_1'], data_class_1['x_2'], 'ro', alpha=.6)

for point, pred in zip(X_binary_class,y_binary_class):
    proj = np.dot(point,W)/np.dot(W,W) * W
    y = np.dot(point,W)
    ax.scatter(proj[0]-1.5,proj[1],color=colors[int(pred)], alpha=0.2)
    
ax.scatter(m0[0],m0[1],color='black',s=100,marker="X")
ax.scatter(m1[0],m1[1],color='black',s=100,marker="X")
plt.legend((0,1))
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.text(1.5, -1, '$X$ = mean', fontsize=13)
plt.text(-1.3,-.5, 'Fisher\'s Discriminant')
plt.show()
```

- Histograma das classes projetadas:

```python
yA = np.squeeze(np.dot(C0,np.expand_dims(W,1)))
yB = np.squeeze(np.dot(C1,np.expand_dims(W,1)))
n_bins = 15
fig, axes = plt.subplots(nrows=1, ncols=1)
colors = ['red', 'blue']
axes.hist([yA,yB], n_bins, density=1, color=colors, histtype='bar', stacked=True)
axes.legend('01')
axes.set_title('Fisher\'s Discriminant Histogram')
plt.show()
```

Como mencionado, devido a simetria do problema, ainda há uma sobreposição dos histogramas referentes as classes. No entanto, há fortes indícios de que esta é a projeção ótima de Fisher.

### 3. Obtendo a curva ROC e F1-Score da LDA

A célula abaixo realiza a soma dos verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos para o limiar (_threshold_) $th = [-0.05, 0.05]$, com 100 pontos. Sequencialmente, realiza o cálculo da **Acurácia** máxima, **F1-Score** e __$th$__ correspondente a essa acurácia.

```python
v_recalls = []
fp_tx = []
v_acc = []
f1_score = []


threshold = np.linspace(-0.05,0.05,100)

for w0 in threshold:
    predictions = []
    true_positives = 0
    false_positives = 0
    true_negatives = 0
    false_negatives = 0

    for pred in y_hat:
        if pred < w0:
            predictions.append(0)
        else:
            predictions.append(1)

    for pred, i in zip(predictions,y_binary_class):
        if pred == 0:
            if pred == i:
                true_negatives +=1
            else:
                false_negatives +=1
        else:
            if pred == i:
                true_positives +=1
            else:
                false_positives +=1

    # Acurácia:
    acc = (true_positives + true_negatives) / len(y_binary_class)

    # Precisão:
    try:
        precision = true_positives / (true_positives + false_positives)
    except ZeroDivisionError:
        precision = 0

    # Taxa de falso positivo:
    FP_tx = false_positives / (true_negatives + false_positives)

    # RECALL: Proporção de amostras da classe positiva corretamente classificadas:
    recall = true_positives / (true_positives + false_negatives)
    
    # F1-Score:
    try:
        f1 = (2*recall*precision)/(recall + precision)
    except ZeroDivisionError:
        f1 = 0
    
    #print("#"*10)
    #print("Threshold:", w0)
    #print("Accurácia:", acc)
    #print("Precisão:", precision)
    #print("Recall:", recall)
    #print("F1-medida:", f1)
    v_recalls.append(recall)
    fp_tx.append(FP_tx)
    v_acc.append(acc)
    f1_score.append(f1)
    
index_max_vacc = v_acc.index(max(v_acc)) # Pega o índice da acurácia máxima
print('Threshold:',threshold[index_max_vacc]) # Threshold correspondente
print('Acurácia:',v_acc[index_max_vacc]) # Acurácia correspondente
print('F1-Score:',f1_score[index_max_vacc]) # F1-Score correspondente
```

Portanto, o valor _threshold_ que maximiza a acurácia é $th = -0.0035$, onde obtemos a acurácia e F1-Score de **95%**

A título de **comparação**, utilizaremos a função de LDA pronta do _Scikit-Learn_ para obtermos a acurácia:

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

clf = LinearDiscriminantAnalysis()
clf.fit(X_binary_class, y_binary_class)
print('# Comparação com Scikit-Learn #')
print("Acurácia:",clf.score(X_binary_class,y_binary_class))
```

Resultando na mesma acurácia do classificador projetado. Desse modo, podemos **afirmar** que o discriminante de projetado é ótimo.

A célula abaixo plota as curvas **ROC** e **F1-Score** para o seguinte conjunto de _thresholds_ $th = [-0.05, 0.05]$, com 100 pontos.

```python
# ROC
plt.figure(1)
plt.plot(fp_tx, v_recalls, 'b-')
plt.grid(True, linestyle='--', linewidth=0.8)
plt.xlabel('False Positive Rate')
plt.gca().legend(('ROC',))
plt.ylabel('True Positive Rate ($Recall$)')
# F1-Score
plt.figure(2)
plt.plot(threshold,f1_score, 'r-')
plt.grid(True, linestyle='--', linewidth=0.8)
plt.xlabel('Threshold ($th$)')
plt.ylabel('F1-Score')
plt.show()    
```

### 4. Classificação Binária com Regressão Logística

Para classificação binária, a regressão logística se dá através do seguinte mapeamento:

$\hat{y}(x)$ = $\dfrac{1}{1 + e^{-(X^Tw)}}$, onde $X$ é a matriz dos padrões (adicionando a primeira coluna de 1s), e $w$ o vetor coluna de pesos ótimos.

$w$ é encontrado através da minimização da função custo $J(w)$ = $-\dfrac{1}{N}$$\sum_{i=0}^{N-1} y_ilog\hat{y}(x(i)) + (1-y_i)log(1-\hat{y}(x(i)))$.

No entanto, por ser uma função convexa, **não admite** solução em forma fechada, tendo que recorrer a algoritmos iterativos. Utilizaremos, portanto, a regressão logísitica com gradiente descendente.

Para esse problema, foi criado um arquivo auxiliar _logisticregression.py_, contendo a classe _LogisticRegressionSGD_ que realiza a classificação binária utilizando a regressão logística iterativa. O _output_ da célula abaixo nos dá o vetor de pesos **$w$** ótimo, bem como o intercepto $w_0$.

```python
from logisticregression import *

X = data.drop('y', axis=1)
y = data['y'].values
N =  X.shape[0] # Número de amostras
n = X.shape[1] # Número de features dos padrões

l_regression = LogisticRegressionSGD()

W, w0 = l_regression.weights(n)

# Gradiente Descendente
coeff, _, _ = l_regression.model_predict(W, w0, X, y, 000.1, 4500)

# Predição final
w = coeff[0]
w0 = coeff[1]
print('Pesos ótimos:', w)
print('w0 (Intercepto) ótimo:', w0)

H = l_regression.sigmoid(np.dot(w,X.T)+w0)

y_hat = l_regression.predict(H, N)
```

- A seguir, é feito o _plot_ da fronteira de decisão juntamente aos padrões no espaço original dos dados

```python
data_class_1 = data.loc[data['y'] == 1] # Dataframe dos padrões da classe 1
data_class_0 = data.loc[data['y'] == 0] # Dataframe dos padrões da classe 0
x = np.linspace(-1.5,2.5,2)
x2 = (-1/w[0][1]) * (w[0][0] * x + w0)

plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data_class_0['x_1'], data_class_0['x_2'], 'o', alpha=.6)
plt.plot(data_class_1['x_1'], data_class_1['x_2'], 'ro', alpha=.6)
plt.plot(x,x2, 'k--')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend((0,1,'Decision boundary'))
plt.show()
```

Análogo ao item anterior, a próxima célula realiza a soma dos verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos para o limiar (_threshold_) $th = [-1.5, 1.5]$, com 100 pontos. Sequencialmente, realiza o cálculo da **Acurácia** máxima, **F1-Score** e __$th$__ correspondente a essa acurácia.

```python
y = np.asarray(data['y'])
v_recalls = []
fp_tx = []
v_acc = []
f1_score = []
threshold = np.linspace(-1.5,1.5,100)

for w0 in threshold:
    predictions = []
    true_positives = 0
    false_positives = 0
    true_negatives = 0
    false_negatives = 0

    for pred in H[0]:
        if pred < w0:
            predictions.append(0)
        else:
            predictions.append(1)

    for pred, i in zip(predictions,y):
        if pred == 0:
            if pred == i:
                true_negatives +=1
            else:
                false_negatives +=1
        else:
            if pred == i:
                true_positives +=1
            else:
                false_positives +=1

    # Acurácia:
    acc = (true_positives + true_negatives) / len(y)

    # Precisão:
    try:
        precision = true_positives / (true_positives + false_positives)
    except ZeroDivisionError:
        precision = 0

    # Taxa de falso positivo:
    FP_tx = false_positives / (true_negatives + false_positives)

    # RECALL: Proporção de amostras da classe positiva corretamente classificadas:
    recall = true_positives / (true_positives + false_negatives)
    
    # F1-Score:
    try:
        f1 = (2*recall*precision)/(recall + precision)
    except ZeroDivisionError:
        f1 = 0
    
    #print("#"*10)
    #print("Threshold:", w0)
    #print("Accurácia:", acc)
    #print("Precisão:", precision)
    #print("Recall:", recall)
    #print("F1-medida:", f1)
    v_recalls.append(recall)
    fp_tx.append(FP_tx)
    v_acc.append(acc)
    f1_score.append(f1)
    
index_max_vacc = v_acc.index(max(v_acc)) # Pega o índice da acurácia máxima
print('Threshold:',threshold[index_max_vacc]) # Threshold correspondente
print('Acurácia:',v_acc[index_max_vacc]) # Acurácia correspondente
print('F1-Score:',f1_score[index_max_vacc]) # F1-Score correspondente
```

Da mesma forma, o valor _threshold_ que maximiza a acurácia é $th$ = $-0.378$, onde obtemos a acurácia e F1-Score de **95%**. A célula seguinte utiliza a função de Regressão Logística pronta do _Scikit-Learn_ para obtermos a acurácia:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 

model = LogisticRegression(solver='liblinear')
model.fit(X, y)
predicted_classes = model.predict(X)
accuracy = accuracy_score(y.flatten(),predicted_classes)
parameters = model.coef_
print('# Comparação com Scikit-Learn #')
print('Parâmetros ótimos:',parameters)
print('Acurácia:',accuracy)
```

Resultando em uma acurácia bem parecida do classificador projetado. Além disso, o vetor de pesos gerado também é muito próximo do encontrado.

A célula abaixo plota as curvas ROC e F1-Score para o seguinte conjunto de _thresholds_ $th$ = $[−1.5, 1.5]$, com 100 pontos.

```python
# ROC
plt.figure(1)
plt.plot(fp_tx, v_recalls, 'b-')
plt.grid(True, linestyle='--', linewidth=0.8)
plt.xlabel('False Positive Rate')
plt.gca().legend(('ROC',))
plt.ylabel('True Positive Rate ($Recall$)')
# F1-Score
plt.figure(2)
plt.plot(threshold,f1_score, 'r-')
plt.grid(True, linestyle='--', linewidth=0.8)
plt.xlabel('Threshold')
plt.ylabel('F1-Score')
plt.show()    
```

Apesar do limiar de decisão ser diferente, as duas formas de classificação (LDA e Regressão Logística) obtiveram praticamente a mesma taxa de acerto, fenômeno que explica as respectivas curvas _ROC_ e _F1-Score_ serem praticamente iguais.
